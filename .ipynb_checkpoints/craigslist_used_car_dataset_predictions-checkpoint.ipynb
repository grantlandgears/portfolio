```python
# copied from above becuase I was running it so much to reset
df_cars = pd.read_csv('./data/final_project/vehicles.csv')
```

```python
# drop columns we don't want to use anywhere in any scenario

## !! df is now the dataset NOT df_cars

df = df_cars.drop(['paint_color', 'state', 'region', 'id', 'county', 'url', 'region_url', 'image_url', 'vin', 'description'], axis=1)
```

```python
# drop rows with null values
df.dropna(inplace=True)
```

```python
# change odometer and year from float to int
df = df.astype({"odometer":'int', "year":'int'})
```

```python
# rename porche to porsche
def rename_val(val):
    if val == 'porche':
        return 'porsche'
    return val

df['manufacturer'] = df['manufacturer'].apply(rename_val)
```

```python
# rename pickup to truck
def rename_val(val):
    if val == 'pickup':
        return 'truck'
    return val

df['type'] = df['type'].apply(rename_val)
```

```python
# remove "cylinders" from values in the cylinder column.  Eg, change "8 cylinders" to just "8"
df['cylinders'] = df.cylinders.str.split(' ').str[0] + ''
```

```python
# remove 191 rows of vehicles who contain 'other' for cylinders
# I looked at the vehicles with 'other' and they were not consistent, some were 8 cylinders, 4 cylindres, and others electric
df = df[df.cylinders != 'other']
```

```python
# change cylinders from object to int
df['cylinders'] = df['cylinders'].astype(int)
```

```python
#DROP DUPES
#df.drop_duplicates(subset=['price','manufacturer','model','lat'], keep="first", inplace=True)

df.drop_duplicates(subset=['price','manufacturer','model', 'odometer'], keep="first", inplace=True)

#df.info()
# check that dupes were dropped by comnparing dataframes before and after
# this required the top call in teh cell to write out a new df called df_no_dups
#df_diff = df.merge(df_no_dups, how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']
#df_diff.sort_values(['price', 'year','manufacturer', 'model', 'odometer'])
```

```python
df.query("model == 'azera limited' & price == 5790")
```

```python
# get rid of cars priced higher than 100000
# there were a couple cars in here that were probably valid, but the majority were bogus

#df.query("price > 100000").sort_values(by=['price'], ascending=False)

df.drop(df[df['price'] > 100000].index, inplace = True)
```

```python
# get rid of cars priced lower than 100
# there were a couple cars in here that were probably valid, but the majority were bogus

df.query("price < 500").sort_values(by=['price'], ascending=False)

df.drop(df[df['price'] < 100].index, inplace = True)
```

```python
# get rid of cars priced lower than 500 AND are listed as good or better condition and mileague under 100k

#df.query("price < 500 & condition != ('salvage', 'fair') & odometer < 100000").sort_values(by=['price'], ascending=False)

df.drop(df[(df['price'] < 500) & (df['condition'] != ('salvage', 'fair')) & (df['odometer'] < 100000)].index, inplace=True)
```

```python
# get rid of cars with less than 100 miles that are NOT new

df.query("odometer < 100 & condition != 'new'").sort_values(by=['odometer'], ascending=False)

df.drop(df[(df['odometer'] < 100) & (df['condition'] != 'new')].index, inplace=True)
```

```python
# get rid of cars with less than 1000 miles that are NOT new OR like new
# there were a couple cars in here that were probably valid, but the majority were bogus

#df.query("odometer < 1000 & condition != ('new', 'like new')").sort_values(by=['odometer'], ascending=False)

df.drop(df[(df['odometer'] < 1000) & (df['condition'] != ('new', 'like new'))].index, inplace=True)
```

```python
# get rid of cars with less than 100 miles that are NOT new
# there were a couple cars in here that were probably valid, but the majority were bogus

#df.query("odometer > 300000").sort_values(by=['odometer'], ascending=False)

df.drop(df[(df['odometer'] > 300000)].index, inplace=True)
```

```python
### !!! YOU NEED TO RUN THIS ON A SUBSET OF DATA.  TRY WITH ~1K ROWS) !!!

# ensure full output from GridSearch
set_config(print_changed_only=False)

# dictionary to hold each hyper parameter and the values to use.  max_features; auto, sqrt, log2 are used below, but
# this dictionary of paramters could be as big as you want (that is any available hyper parameter)
# here are all the combinations I want grid search to to try. For example: 200, auto, 4, gini THEN 200, sqrt, 4, gini, etc...
# at the end it will return the model that did the best
param_grid = {
    'n_estimators': [20, 50],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [4,8],
    # gini and entropy usually are similar
    'criterion':['mse', 'mae']
}

# instatiate the random forest classifier (rf)
rf = RandomForestRegressor()

# cross fold (cv) like you did with kfold.  Train on four folds test on 1 (4+1=5)
# run through all options outlined in the param_grid above
# we are doing the same steps of an instantied model where we fit, predict, and score
cv_rf = GridSearchCV(rf, param_grid=param_grid, cv= 3)
cv_rf.fit(Xt_train, yt_train)

# the output of all this (cv_rf) is a new model just like any of the five three cells up.
```

```python
# you can now do all the normal steps on a model
# show you the best model that was found

cv_rf.best_estimator_
```

```python
# what was the best score

cv_rf.best_score_
```

```python
# now that we have the best estimator we can run it on test data
best_train = cv_rf.best_estimator_.predict(Xt_train)
best_test = cv_rf.best_estimator_.predict(Xt_test)
```
